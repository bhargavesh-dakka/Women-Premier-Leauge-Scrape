```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
import warnings
warnings.simplefilter("ignore")
```

#READ AS --> Player Name - Base Price (in ₹) - Selling Price (in ₹) - Team

```python
link = "https://sportstar.thehindu.com/cricket/wpl-auction-2023-top-sold-unsold-players-list-live-streaming-updates-wipl-womens-ipl/article66502635.ece"
```

```python
response = requests.get(link)
response.status_code
```

```python
wpl = BeautifulSoup(response.content, "html.parser")
```

```python
data = wpl.find_all("div", class_ = "ng_infobox factbox-container")
```

```python
data_dict = dict()
for content in data:
    S = content.find("div").string
    
    d = content.find_all("div", class_ = "infobox-description")
    if S in data_dict:
        S = "Unsold" + S
    data_dict[S] = []
    for datum in d:
        data_dict[S].append(datum.string)
        
        
    
    
```

```python
data_dict
```

```python
WPL = pd.DataFrame(columns =["Name", "Base Price (rs)", "Sold Price (rs)", "Team"])
```

```python
WPL
```

```python

for column in data_dict:
    for d in data_dict[column]:
        name.append(d.split("-")[0])
        base.append(d.split("-")[1])
        try:
            sold.append(d.split("-")[2])
        except:
            sold.append("Unsold")
        try:    
            team.append(d.split("-")[3])
        except:
            team.append("Unsold")
        
        
```

```python
name = []
base = []
sold = []
team = []
for column in data_dict:
    for d in data_dict[column]:
        name.append(d.split("-")[0])
        base.append(d.split("-")[1])
        try:
            sold.append(d.split("-")[2])
        except:
            sold.append("Unsold")
        try:    
            team.append(d.split("-")[3])
        except:
            team.append("Unsold")
        
        
```

```python
WPL["Name"] = pd.Series(name)
WPL["Base Price (rs)"] = pd.Series(base)
WPL["Sold Price (rs)"] = pd.Series(sold)
WPL["Team"] = pd.Series(team)
```

```python
for column in WPL:
    print(column)
    WPL[column] = WPL[column].apply(lambda X : X.replace("\n",""))
```

```python
for column in WPL:
    if column not in ("Name","Team"):
        print(column)
        WPL[column] = WPL[column].apply(lambda X : X.replace(r" lakh","00000"))
```

```python
for column in WPL:
    if column not in ("Name","Team"):
        print(column)
        
        WPL[column] = WPL[column].apply(lambda X : X.replace(r" crore","000000") if "." in X else X.replace(r" crore","0000000") )
        WPL[column] = WPL[column].apply(lambda X : X.replace(r".",""))
```

```python
WPL.to_csv("Women Premier Leauge.csv")
```
